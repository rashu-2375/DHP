convert these numbers to z-score(in a  new dataframe) and redo the box plot



find mew, and sigma using pandas from describe

```python
desc = df.describe()
desc
```

```python
desc['tifflers']['mean']
```

```python
new_df = pd.DataFrame()
for col in df:
    if col == "Date" :
        new_df['Date']=df['Date']
    else:
        new_df[col]= (df[col].values - desc[col]['mean'])/desc[col]['std']

```

```python
new_df
```

```python
df_new_melt = pd.melt(new_df.drop("Date", axis = 1))
df_new_melt
```

```python
fig=px.box(df_new_melt, x="variable", y="value", points="all",color="variable")
fig.show()
```

```python
mean_fancave = df["fancave"].mean()
```

```python
mean_fancave
```

```python
std_fancave = df["fancave"].std()
```

```python
std_fancave
```

```python
z_scores_fancave = (df["fancave"] - mean_fancave) / std_fancave
```

```python
z_scores_fancave
```

```python
z_scores_df = pd.DataFrame({'fancave_zscore': z_scores_fancave})
```

```python
print(z_scores_df)
```

```python
fig_zscores = go.Figure()
fig_zscores.add_trace(go.Box(y=z_scores_df['fancave_zscore'], name='fancave_zscore'))
fig_zscores.show()
```

```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Calculate z-scores for numeric columns only
z_scores_df = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()

# Create a box plot for all z-scores
fig_zscores = go.Figure()

for col in z_scores_df.columns:
    fig_zscores.add_trace(go.Box(y=z_scores_df[col], name=f'{col}_zscore'))

fig_zscores.show()
```

Heatmap

```python
stock_names  = ["fancave", "tifflers", "canteen"]
stock_value_start=[df.iloc[0][stock] for stock in stock_names]
fig=px.pie(names=stock_names, values=stock_value_start)
fig.show()
```

```python
stock_value_end = [df.iloc[-1][stock] for stock in stock_names]
fig=px.pie(names=stock_names, values=stock_value_end)
fig.show()
```

subplots

```python
fig = make_subplots(rows=1, cols=2,specs=[[{"type": "domain"}, {"type": "domain"}]], subplot_titles=("Start Valuation","End Valuation"))
fig.add_trace(
    go.Pie(labels=stock_names, values=stock_value_start, domain=dict(x=[0, 0.5])),
    row=1, col=1)
fig.add_trace(
    go.Pie(labels=stock_names, values=stock_value_end, domain=dict(x=[0.5, 1.0])),
    row=1, col=2)
fig.update_layout(height=600, width=800, title_text="Stock Valuations")
fig.show()
```
