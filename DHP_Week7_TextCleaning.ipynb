# Regular Expressions

Both chatGPT and Gemini are very bad with Python Regular Expressions!

```python
import re  # regexp, are powerful tools for pattern matching and manipulation of strings. 

# findall	: Returns a list containing all matches
# search	: Returns a Match object if there is a match anywhere in the string
# split	: Returns a list where the string has been split at each match
# sub	: Replaces one or many matches with a string
```

```python
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

```python
# Find words ending with "ed".
# The symbol $ refers to the end of the string

# [w for w in wordlist if re.search('ed$', w)]
[w for w in wordlist if re.search('ed', w)]
```

```python
# The . wildcard symbol matches any single character.
# Suppose we have room in a crossword puzzle for an 8-letter word with j as its third letter and t as its sixth letter.
# In place of each blank cell we use a period, and the symbol ^ matches the beginning of the sentence:

[w for w in wordlist if re.search('^..j..t..$', w)]
```

**DIY:**

 What results do we get with the above example if we leave out both ^ and $, and just search for «..j..t..»?

```python
[w for w in wordlist if re.search('..j..t..', w)]
```

```python
# The ? symbol specifies that the previous character is optional.
# Thus «^be?ad$» will match both bad and bead.

[w for w in wordlist if re.search('^be?ad$', w)]
# [w for w in wordlist if re.search('^be?.d$', w)]
```

```python
# Whats happening here?
#[ghi]: Matches any single character among 'g', 'h', or 'i'.
# [w for w in wordlist if re.search('^[ghi]$', w)]
[w for w in wordlist if re.search('^ghi$', w)]
        #looks for words that consist of the exact sequence 'ghi'.
```

```python
# Whats happening here?
#looks for words that are exactly four characters long, 
#with each character being from the specified sets.
[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
```

```python
# Whats happening here?
# looking for strings that consist entirely of one or more occurrences of 
#the letters 'g', 'h', or 'i'. The + sign indicates that at least one of these letters must be present, but there can be more.
[w for w in wordlist if re.search('^[ghi]+$', w)]
```

```python
# Whats happening here?
# range [g-i] allows for flexibility in matching any combination of these letters.
[w for w in wordlist if re.search('^[g-i]+$', w)]
```

```python
# Whats happening here?

[w for w in wordlist if re.search('^[g-j]+$', w)]
```

```python
# Whats happening here?
#  excludes strings that contain any characters from 'c' to 'z'.
[w for w in wordlist if re.search('^[^c-z]+$', w)]
```

Characters that are not within a range can be matched by complementing the set. If the first character of the set is '^', all the characters that are not in the set will be matched. For example, [^5] will match any character except '5', and [^^] will match any character except '^'. ^ has no special meaning if it’s not the first character in the set.

# Remove Punctuations

```python
# initializing string
test_str = "Gfg, is best : for ! Geeks ;"

# printing original string
print("The original string is : \n" + test_str)
print("\n")

# Removing punctuations in string
# Using regex
# \w: Represents any word character (alphanumeric plus underscore).
# \s: Represents any whitespace character (spaces, tabs, newline).
res = re.sub(r'[^\w\s]', '', test_str)

# printing result
print("The string after punctuation filter : \n" + res)
```

Characters that are not within a range can be matched by complementing the set. If the first character of the set is '^', all the characters that are not in the set will be matched. For example, [^5] will match any character except '5', and [^^] will match any character except '^'. ^ has no special meaning if it’s not the first character in the set.

# Unicode Encodings

**Unicode** supports over a million characters. Each character is assigned a number, called a code point. In Python, code points are written in the form \uXXXX, where XXXX is the number in 4-digit hexadecimal form.



image.png

```python
path = nltk.data.find('corpora/unicode_samples/polish-lat2.txt')
```

```python
f = open(path, encoding='latin2')
for line in f:
  line = line.strip()
  print(line)
```

```python
f = open(path, encoding='latin2')
for line in f:
  line = line.strip()
  print(line)

  encoded_text = line.encode('unicode_escape')
  print(encoded_text)

  ascii_text = line.encode('ascii','ignore')
  print(ascii_text)

  print("\n")
```

# GENSIM

Source : https://www.kaggle.com/code/venkatkrishnan/gensim-text-mining-techniques

```python
import re
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# pd.set_option('display.max_colwidth', -1)

# dataset
from sklearn.datasets import fetch_20newsgroups

# Gensim packages
import gensim
from gensim.parsing import strip_tags, strip_numeric, strip_multiple_whitespaces, stem_text, strip_punctuation, remove_stopwords
from gensim.parsing import preprocess_string
```

```python
# loading dataset
news_group = fetch_20newsgroups(subset='train')

news_group_data = news_group.data
news_group_target_names = news_group.target_names
news_group_target = news_group.target
```

```python
print(len(news_group_data),len(news_group_target_names),len(news_group_target))
```

```python
print(news_group_target_names)
```

```python
# Creating a dataframe from the loaded data
news_df = pd.DataFrame({'news': news_group_data,
                        'class': news_group_target})
```

```python
news_extracts = news_df.sample(2000)

news_extracts.reset_index(drop=True, inplace=True)
```

```python
news_extracts.head(2)
```

```python
# Custom filter method
transform_to_lower = lambda s: s.lower()
remove_single_char = lambda s: re.sub(r'\s+\w{1}\s+', '', s)
# Filters to be executed in pipeline
CLEAN_FILTERS = [strip_tags,
                strip_numeric,
                strip_punctuation,
                strip_multiple_whitespaces,
                transform_to_lower,
                remove_stopwords,
                remove_single_char]

# Method does the filtering of all the unrelevant text elements
def cleaning_pipe(document):
    # Invoking gensim.parsing.preprocess_string method with set of filters
    processed_words = preprocess_string(document, CLEAN_FILTERS)

    return processed_words
```

```python
# Apply the cleaning pipe on the news data

news_extracts['clean_text'] = news_extracts['news'].apply(cleaning_pipe)
```
